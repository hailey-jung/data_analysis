```python
# 인공신경망(미분) - 경사하강법

import numpy as np

x1 = np.array([2, 4, 6, 8])
x2 = np.array([1, 2, 1, 3]) # 문제집 푼 개수수
y = np.array([25, 60, 55, 90])

a1 = 0
a2 = 0
b = 0
lr = 0.005

for n in range(3000):
    y_hat = a1 * x1 + a2 * x2 + b
    error = y_hat - y
    
    a1_diff = sum(2 * error * x1)
    a2_diff = sum(2 * error * x2)
    b_diff = sum(2 * error * 1)

    a1 = a1 - a1_diff * lr
    a2 = a2 - a2_diff * lr
    b = b - b_diff * lr

    print("학습 {}회, a1 : {}, a2 : {}, b : {}".format(n+1, a1, a2, b))
```

```python
plt.scatter(x1, y)
plt.plot(x1, a1 * x1 + a2 * x2 + b, 'red')
```

```python
import random

a = random.random()
b = random.random()
c = random.random()
d = random.random()

lr = 0.000001

for n in range(5000):
    y_hat = a * x ** 3 + b * x ** 2 + c * x + d
    error = y_hat - y

    a_diff = sum(2 * error * x ** 3)
    b_diff = sum(2 * error * x ** 2)
    c_diff = sum(2 * error * x)
    d_diff = sum(2 * error * 1)

    a = a - a_diff * lr
    b = b - b_diff * lr
    c = c - c_diff * lr
    d = d - d_diff * lr

    print("{}회 학습, a : {}, b : {}, c : {}, d : {}".format(n+1, a, b, c, d))
```

```python
# 정형 데이터 : 트리 기반(분할X, 표준화X)
# 결정트리 -> 랜덤포레스트, XGboost(O)
# 비정형 데이터 : 딥러닝

from sklearn.tree import DecisionTreeClassifier, plot_tree

# 스무고개
dt = DecisionTreeClassifier()
dt.fit(train_input, train_target)

# 과대적합 : 훈련>>>>>시험
# 과소적합 : 훈련<시험, 둘다 낮은 경우

print(dt.score(train_input, train_target))
print(dt.score(test_input, test_target))
```
