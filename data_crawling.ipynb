```python
# 데이터 크롤링

# 파이썬으로 HTML을 다루는 모듈
# 페이지를 요청하는 모듈
# 엑셀화 모듈
# 시간 모듈

from bs4 import BeautifulSoup # Python으로 HTML을 다루는 기능능
import requests # 페이지 요청
import pandas as pd
import time
```

```python
# 200 : 정상, 400번대 : 존재하지않는페이지, 500번대 : 거절
url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EB%A1%9C%EB%98%90")
html = BeautifulSoup(url.text)

current = int(html.find("span", class_ = 'f_red').text.replace("회", ""))

number = html.find("div", class_ = 'lottonum').find_all("span")

del number[-2]
del number[-2]

box = []
for i in number:
    box.append(int(i.text))
box
```

```python
total = []

for n in range(1, 11):
    url = requests.get(f"https://search.daum.net/search?w=tot&DA=LOT&rtmaxcoll=LOT&&q={n}%ED%9A%8C%EC%B0%A8%20%EB%A1%9C%EB%98%90")
    html = BeautifulSoup(url.text)

    number = html.find("div", class_ = 'lottonum').find_all("span")

    del number[-2]
    del number[-2]

    box = []
    for i in number:
        box.append(int(i.text))
    
    total.append(box)
    print(f"{n}회 로또 데이터 저장완료... {box}")
    time.sleep(1)
```

```python
total
```

```python
df = pd.DataFrame(total, 
                  columns = ['1번째','2번째','3번째','4번째','5번째','6번째','보너스'])
df.to_excel("lotto.xlsx")
df
```

```python
# 연금복권 데이터 크롤링하기(1초씩 쉬기기)

url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C")
html = BeautifulSoup(url.text)

current = int(html.find("span", class_ = 'f_red').text[1:-1])

number = html.find("tr", class_ = 'fst').text.split()[5:]
number[0] = number[0].replace("조", "")
number = list(map(int, number)) # number안에 모든 요소에 int함수를 적용해라라
number
```

```python
total = []
for n in range(1, current + 1):
    url = requests.get(f"https://search.daum.net/search?w=tot&DA=JIM&rtmaxcoll=JIM&&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C%20720%2B%20{n}%ED%9A%8C%EC%B0%A8")
    html = BeautifulSoup(url.text)
    number = html.find("tr", class_ = 'fst').text.split()[5:]
    number[0] = number[0].replace("조", "")
    number = list(map(int, number))
    total.append(number)
    time.sleep(1)
    print(f"{n}회 연금복권 : {number}")
```

```python
df = pd.DataFrame(total,
                  columns = ['조','1번째','2번째','3번째','4번째','5번째','6번째'])
df.to_excel("pension.xlsx")
df
```

```python
from tqdm import tqdm # for문의 진행상황 확인인

url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
html = BeautifulSoup(url.text)
kospi_page = int(html.find('td', class_ = 'pgRR').find('a')['href'][-2:])

total = []
for n in tqdm(range(1, kospi_page + 1)):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={n}")
    html = BeautifulSoup(url.text)
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0]
    table = table[table['종목명'].notnull()]
    del table['N']
    del table['토론실']
    total.append(table)
    time.sleep(1)
kospi = pd.concat(total, ignore_index=True)
kospi.to_excel("kospi.xlsx")

#--

url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1")
html = BeautifulSoup(url.text)
kosdaq_page = int(html.find('td', class_ = 'pgRR').find('a')['href'][-2:])

total = []
for n in tqdm(range(1, kosdaq_page + 1)):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={n}")
    html = BeautifulSoup(url.text)
    table = html.find('table', class_ = 'type_2')
    table = pd.read_html(str(table))[0]
    table = table[table['종목명'].notnull()]
    del table['N']
    del table['토론실']
    total.append(table)
    time.sleep(0.1)
kosdaq = pd.concat(total, ignore_index=True)
kosdaq.to_excel("kosdaq.xlsx")
```

```python
kospi['소속'] = 'KOSPI'
kosdaq['소속'] = "KOSDAQ"
```

```python
stock = pd.concat([kospi, kosdaq], ignore_index = True)
stock.to_excel("stock.xlsx")
stock
```
